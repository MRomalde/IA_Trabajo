```python
import networkx as nx
import matplotlib.pyplot as plt
graph = nx.Graph()  
graph.add_node('you')
graph.add_node('me')
graph.add_edge('me','you')
print(graph.number_of_edges())
print(graph.number_of_nodes())
print(list(graph.nodes))
print(list(graph.edges))
print(list(graph.neighbors('you')))#number of neighbors of node 'you'
print(graph.degree('you'))# the number of edges incident to 1
#graph.remove_node('me')#borrar 1 nodo
#graph.remove_edge('me','you')#borrar 1 edge
```

```python
edgelist = [(0, 1), (1, 2), (2, 3)]
H = nx.Graph(edgelist)  # create a graph from an edge list
print(list(H.edges()))
```

```python
adjacency_dict = {0: (1, 2), 1: (0, 2), 2: (0, 1)}
H = nx.Graph(adjacency_dict)  # create a Graph dict mapping nodes to nbrs
print(list(H.edges()))
```

```python
FG = nx.Graph()
FG.add_weighted_edges_from([(1, 2, 0.125), (1, 3, 0.75), (2, 4, 1.2), (3, 4, 0.375)])
for n, nbrs in FG.adj.items():
   for nbr, eattr in nbrs.items():
       wt = eattr['weight']
       if wt < 0.5: print(f"({n}, {nbr}, {wt:.3})")
```
